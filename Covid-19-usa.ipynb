```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import urllib.request
from urllib.request import urlopen
import requests
import re
import os
from geopy.geocoders import Nominatim
import folium
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
url='https://www.worldometers.info/coronavirus/country/us/'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
tables=soup.find_all('table',{'class':'table table-bordered table-hover table-responsive usa_table_countries'})
tables
```

```python
states=[]
cases=[]
deaths=[]
active=[]
for table in tables:
    rows=table.find_all('tr')
    for row in rows:
        cells=row.find_all('td')
        if(len(cells)>1):
            states.append(cells[0].text.strip())
            cases.append(cells[1].text.strip())
            deaths.append(cells[3].text.strip())
            active.append(cells[5].text.strip())
```

```python
dataset={'State':states,'Total Cases':cases,'Total Deaths':deaths,'Total Active Cases':active}
df=pd.DataFrame(data=dataset)
```

```python
drop=list(range(61,df.shape[0]))
a2=[60,30,58,52,57,53,54,56,0,55,59]
for element in a2:
    drop.append(element)
drop
```

```python
df.drop(drop,axis=0,inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
address='USA'
geolocator = Nominatim(user_agent="us_explorer")
location = geolocator.geocode(address)
us_lat = location.latitude
us_lng = location.longitude
print(us_lat,us_lng)
```

```python
for i in range(0,df.shape[0]):
    cell1=df.loc[i,'Total Cases']
    cell_fixed1=''
    cell2=df.loc[i,'Total Deaths']
    cell3=df.loc[i,'Total Active Cases']
    cell_fixed2=''
    cell_fixed3=''
    for char in cell1:
        if (char==','):
            continue
        else:
            cell_fixed1=cell_fixed1+char
    for char in cell2:
        if (char==','):
            continue
        else:
            cell_fixed2=cell_fixed2+char
    for char in cell3:
        if (char==','):
            continue
        else:
            cell_fixed3=cell_fixed3+char
    
    df.at[i,'Total Cases']=cell_fixed1
    df.at[i,'Total Deaths']=cell_fixed2
    df.at[i,'Total Active Cases']=cell_fixed3
df
```

```python
df=df.astype({'Total Cases':'float64','Total Deaths':'float64','Total Active Cases':'float64'})
```

```python
url='https://www.50states.com/abbreviations.htm'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
tables=soup.find_all('table',{'class':'spaced stripedRows abbreviations-table table table-hover'})
tables
```

```python
state=[]
abbreviation=[]
for table in tables:
    rows=table.find_all('tr')
    for row in rows:
        cells=row.find_all('td')
        if(len(cells)>1):
            state.append(cells[0].text.strip())
            abbreviation.append(cells[1].text.strip())
datadict={'State':state,'Abbreviation':abbreviation}
df_abb=pd.DataFrame(data=datadict)
```

```python
url='https://www.latlong.net/category/states-236-14.html'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html.parser')
tables=soup.find_all('table')
tables
```

```python
state=[]
latitude=[]
longitude=[]
for table in tables:
    rows=table.find_all('tr')
    for row in rows:
        cells=row.find_all('td')
        if(len(cells)>1):
            state.append(cells[0].text.strip())
            latitude.append(cells[1].text.strip())
            longitude.append(cells[2].text.strip())
datadict={'State':state,'Latitude':latitude,'Longitude':longitude}
df_latlng=pd.DataFrame(data=datadict)
df_latlng
```

```python
for i in range(0,df.shape[0]):
    cell=df_latlng.loc[i,'State']
    fin=(cell.split(","))[0]
    df_latlng.at[i,'State']=fin
df_latlng 
```

```python
df_final=pd.merge(df_final,df_latlng,on='State')
df_final
```

```python
df_ny=df_final.iloc[0:2,:]
df_rest=df_final.iloc[2:,:]
```

```python
us_map=folium.Map(location=[us_lat,us_lng],zoom_start=4)
us_map
```

```python
state_geo = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json'
```

```python
bins = list(df_final['Total Cases'].quantile([0, 0.25, 0.5, 0.75, 1]))
```

```python
threshold_scale=np.linspace(0,df_ny['Total Cases'].max(),6,dtype=int)
threshold_scale = threshold_scale.tolist()
threshold_scale[-1] = threshold_scale[-1] + 1

folium.Choropleth(geo_data=state_geo,
                  name='choropleth',
                  data=df_ny,
                  columns=['State','Total Cases'],
                  key_on='feature.properties.name',
                  threshold_scale=threshold_scale,
                  fill_color='YlOrRd',
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  legend_name='Total Cases').add_to(us_map)

folium.Choropleth(geo_data=state_geo,
                  name='choropleth',
                  data=df_rest,
                  columns=['State','Total Cases'],
                  key_on='feature.properties.name',
                  fill_color='YlGn',
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  legend_name='Total Cases').add_to(us_map)
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng,cases,deaths,state in zip(df_final.Latitude, df_final.Longitude, df_final['Total Cases'], df_final['Total Deaths'],df_final['State']):
    label="{} Cases:{} , Deaths:{}".format(state,cases,deaths)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        fill=True,
        fill_color='#0032FF',
        fill_opacity=0.7
        ).add_to(us_map)

us_map
```
